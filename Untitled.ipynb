Let's create some synthetic data on the relationship 
$$y = \sin(x_1) + \cos(x_0) + 2x_1^2 + x_0 \cdot \mathrm{e}^{x_1}$$


```python
x0 = np.arange(-1, 1, 0.1)
x1 = np.arange(-1, 1, 0.1)
x0, x1 = np.meshgrid(x0, x1)
y_truth = np.sin(x1) + np.cos(x0) + 2 * x1*x1 + x0 * np.exp(x1)
```

```python
type(y_truth), np.shape(y_truth)
```

```python
font_format = {'family':'Times New Roman', 'size':16}

ax = plt.figure().gca(projection='3d')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
surf = ax.plot_surface(x0, x1, y_truth, rstride=1, cstride=1, 
                       color='g', alpha=0.8)
plt.title('Original Data', fontdict=font_format)
plt.show()
```

Create random training and test data that lies on this surface too

```python
def createData(key):
    if key == 'train':
        rng = np.random.RandomState(0) # random number generator
        X = rng.uniform(-1, 1, 100).reshape(50, 2)
        y = np.sin(X[:, 1]) + np.cos(X[:, 0]) + 2 * np.square(X[:, 1]) + \
            X[:, 0] * np.exp(X[:, 1])
    elif key == 'test':
        rng = np.random.RandomState(1) # random number generator
        X = rng.uniform(-1, 1, 100).reshape(50, 2)
        y = np.sin(X[:, 1]) + np.cos(X[:, 0]) + 2 * np.square(X[:, 1]) + \
            X[:, 0] * np.exp(X[:, 1])
    else:
        raise('You put a wrong key in createData function!')
        
    return X, y


# training dataset
X_train, y_train = createData('train')
X_test, y_test = createData('test')

print("train:", y_train)
print('test:', y_test)
```

## Test for 1D array data

$y = 2*x_0^2$

```python
def createData(key):
    if key == 'train':
        rng = np.random.RandomState(0) # random number generator
        X = rng.uniform(-1, 1, 100)
        y = 0.0008*np.square(X)
    elif key == 'test':
        rng = np.random.RandomState(1) # random number generator
        X = rng.uniform(-1, 1, 100)
        y = 0.0008*np.square(X)
    else:
        raise('You put a wrong key in createData function!')
        
    return X, y


# training dataset
X_train, y_train = createData('train')
X_test, y_test = createData('test')

est_gp = SymbolicRegressor(population_size=200, 
                          generations=300, stopping_criteria=0.0002,
                          p_crossover=0.6, p_subtree_mutation=0.15,
                          p_hoist_mutation=0.05, p_point_mutation=0.1,
                          max_samples=0.9, verbose=1,
                          parsimony_coefficient=0.01, random_state=0)
# The sum of p_crossover, p_subtree_mutation, p_hoist_mutation and 
# p_point_mutation should total to 1.0 or less.
X_train_new = X_train[:, np.newaxis]
est_gp.fit(X_train_new, y_train)
```

Create a symbolic regressor
- Create own function

```python
def mul_exponential(x1, x2):
    f = x1 * np.exp(x2)
    
    return f


mul_exponential = make_function(function=mul_exponential, name='mul_ex', arity=2)
```

```python
function_set_tuple = ('add', 'mul', 'cos', 'sin', mul_exponential)
est_gp = SymbolicRegressor(population_size=200, function_set=function_set_tuple, 
                          generations=300, stopping_criteria=0.2,
                          p_crossover=0.6, p_subtree_mutation=0.15,
                          p_hoist_mutation=0.05, p_point_mutation=0.1,
                          max_samples=0.9, verbose=1,
                          parsimony_coefficient=0.01, random_state=0)
# The sum of p_crossover, p_subtree_mutation, p_hoist_mutation and 
# p_point_mutation should total to 1.0 or less.
est_gp.fit(X_train, y_train)
```

```python
converter = {
    'sub': lambda x, y : x - y,
    'div': lambda x, y : x / y,
    'mul': lambda x, y : x * y,
    'add': lambda x, y : x + y,
    'neg': lambda x    : - x,
    'pow': lambda x, y : x ** y,
    'sin': lambda x    : sympy.sin(x),
    'cos': lambda x    : sympy.cos(x),
    'inv': lambda x    : 1 / x,
    'sqrt': lambda x   : x ** 0.5,
    'mul_ex': lambda x, y : x * sympy.exp(y)
}
```

```python
print(est_gp)
# print('R2:', est_gp.score(X_test, y_test))
next_e = sympy.sympify((est_gp._program), locals=converter)
print('Expression:', next_e)
```
